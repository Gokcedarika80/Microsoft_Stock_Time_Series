```python
# Gerekli kütüphaneler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
df=pd.read_csv("Microsoft_Stock.csv")
series=df["Close"]
#datetime formuna çevirme
df['Date']=pd.to_datetime(df['Date'])
df.set_index('Date',inplace=True)
##günlük log getiri hesaplayalım
df['log_return']=np.log(df['Close'])/np.log(df['Close'].shift(1))
df["log_return"].plot()
plt.show()
df.dropna()
series=df["log_return"]

series_clean=df["log_return"].replace([np.inf,-np.inf],np.nan).dropna()
```

```python
plt.figure(figsize=(10,5))
plot_acf(series_clean,lags=40)
plt.title("Autocorrelation Function")

plt.show()
```

```python
plt.figure(figsize=(10,5))
plot_pacf(series_clean,lags=40,method="ywm")
plt.title("Partial Autocorrelation Function")

plt.show()
```

```python
result=adfuller(series_clean)
print("ADF Statistic: %f" % result[0])
print("p-value: %f" % result[1])
```

```python
train=int(len(df)*0.8)
train,test=series_clean[0:train],series_clean[train:]
```

```python
model=ARIMA(train,order=(1,1,1))
result=model.fit()
print(result.summary())
```

```python
forecast=result.forecast(steps=len(test))
plt.figure(figsize=(10,5))
plt.plot(train.index,train,label="eğitim")
plt.plot(test.index,test,label="test")
plt.plot(test.index,forecast,label="tahmin")
plt.title("ARIMA(1,1,1) BACKtest")
plt.legend()
plt.show()
```

```python
final_model=ARIMA(df['log_return'],order=(1,1,1))
final_fit=final_model.fit()
print(final_fit.summary())

forecast_steps=60
forecast_result=final_fit.get_forecast(steps=forecast_steps)
pred_mean=forecast_result.predicted_mean
conf_int=forecast_result.conf_int()

last_date=df.index[-1]
future_dates=pd.date_range(start=last_date+pd.Timedelta(days=1),periods=forecast_steps,freq="B")

pred_mean.index=future_dates
conf_int.index=future_dates

plt.figure(figsize=(10,5))
plt.plot(df.index,df["log_return"],label="log fiyat")
plt.plot(pred_mean.index,pred_mean,label="tahmin")
plt.fill_between(conf_int.index,
                 conf_int.iloc[:,0],
                 conf_int.iloc[:,1],alpha=0.3,label="güven aralığı")
plt.title("ARIMA(1,1,1)tahmin")
plt.legend()
plt.show()
```
